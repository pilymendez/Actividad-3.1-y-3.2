```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from funpymodeling.exploratory import freq_tbl
import scipy.special as special 
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
```

```python
df.info()
```

### Botón correcto - Administrador

```python
#Declaramos las variables dependientes e independietes para la regresión NO lineal
var_indep1=df[['Administrador']]
var_dep1=df['botón correcto']

#Redefinicmos las variables
x1= var_indep1
y1= var_dep1

#Generamos una función para el modelo mátematico 
def func1 (x, a, b, c):
    return  a*x**2 + b*x +c

#Ajsutamos los parametros de la cunción curve_fit
parametros1, covs = curve_fit(func1, df['Administrador'], df['botón correcto'])
print(parametros1)

#Predicciones
y_pred1 = func1(x1, *parametros1)
```

```python
#Graficamos las predicciones y los datos originales para realizar la comparación
plt.plot(x1, y1, 'bo', label="y-original")
plt.plot(x1, y_pred1, label="y-predecida con y=a*x^2+b*x+c")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid(True)
plt.show()
```

```python
#Calculamos el coeficiente de determinación del modelo
R2_Modelo1 = r2_score(y1, y_pred1) 
print(R2_Modelo1)

#Calculamos el coeficiente de correlación del modelo
R1= np.sqrt(R2_Modelo1)
print(R1)
```

## Tiempo de interacción, color presionado

```python
#Declaramos las variables dependientes e independietes para la regresión NO lineal
var_indep=df1[['color presionado']]
var_dep=df1['tiempo de interacción']

#Redefinicmos las variables
x= var_indep
y= var_dep

#Generamos una función para el modelo mátematico 
def func (x, a, b, c):
    return  a*x**2 + b*x +c

#Ajsutamos los parametros de la cunción curve_fit
parametros, covs = curve_fit(func, df1['color presionado'], df1['tiempo de interacción'])
print(parametros)

#Predicciones
y_pred = func(x, *parametros)
```

```python
#Graficar la Regresión No lineal 
#Graficamos las predicciones y los datos originales para realizar la comparación
plt.plot(x, y, 'bo', label="y-original")
plt.plot(x, y_pred, label="y-predecida con y=a*x^2+b*x+c")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid(True)

#Gráficar la regresión Lineal 
from sklearn.linear_model import LinearRegression 
model = LinearRegression()

model.fit(X=var_indep, y=var_dep)

model.__dict__

#Predecimos los valores de total de accidentes a partir de la variable 
y_pred_lineal = model.predict(X=df1[['color presionado']])
y_pred_lineal

# Línea del modelo lineal (valores predichos)
plt.plot(var_indep, y_pred_lineal, color='red', label='Regresión lineal', linewidth=2)

# Etiquetas y título
plt.xlabel('Color presionado')
plt.ylabel('Tiempo de inteacción')
plt.title('Correlación')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
